```python
events=pd.read_csv('C:\\Users\\aksha\\Desktop\\dash\\events.csv')
```

```python
events.head
```

```python
events.columns
```

```python
events['event_count']=events.ix[:,2:].sum(axis=1)
```

```python
events.set_index(['year','state','drought','flooding','freeze','severe storm','tropical cyclone','wildfire','winter storm'])
```

```python
events.head()
```

```python
t=events.groupby(['State Name','state']).agg({'event_count':'sum'})
```

```python
t.head()
```

```python
t.to_csv('C:\\Users\\aksha\\Desktop\\dash\\eventcount.csv')
```
